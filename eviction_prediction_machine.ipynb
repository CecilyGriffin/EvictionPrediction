## Linear Regression - Evictions and Race

```python

X = renamed_df["num_evictions"].values.reshape(-1, 1)
y = renamed_df["pct_minority"].values.reshape(-1, 1)
print("Shape: ", X.shape, y.shape)
```

```python
plt.scatter(X, y)
plt.xlabel("Number of Evictions")
plt.ylabel("% Minority Population")
plt.savefig('minoritypopulationvseviction.png')
```

```python
# Splitting data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Creating model

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
# Calculate the mean_squared_error and the r-squared value for the testing data
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (RS ): {r2}")
```

```python
# Showing r2 score
model.score(X_test, y_test)
```

## Multivariate Linear Analysis

Since single demographic factors seem to have some relationship to evictions, but not espeically strong ones, a multivariate linear analysis might be more effective in showing certain patterns. We know that the following demographic factors have some relationship to the number of evictions because of our previous analysis. 

```python
# Assign the data to X and y
y = renamed_df["num_evictions"].values.reshape(-1, 1)
X = renamed_df[["pct_minority", "pct_single_female_children"]]

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python

# Plot the Residuals for the Training and Testing data

### BEGIN SOLUTION
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue",
            label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
### END SOLUTION
```
