```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import gmaps
```

```python
#read file
zip_tract = "Resources/ziptract2010.csv"
tract_us = "Resources/2009-2013_American_Community_Survey_Basic_Census_Tract_Data.csv"
```

```python
# Read the data 
zip_tract = pd.read_csv(zip_tract)
tract_us = pd.read_csv(tract_us)
```

```python
merged_df= zip_tract.merge(tract_us, how='right', on= 'TRACT')
merged_df
```

```python
merged_df= merged_df.dropna(how='any')
merged_df
```
